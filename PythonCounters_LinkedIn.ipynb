```python
#Importing Libraries:

from collections import Counter
from datetime import datetime
```

```python
# Printing the text:

print('Enter a line of text: ')
text= input('').lower()
```

```python
# Importing library for regular expressions to clean the text from any punctuation marks or whitespaces:

import re
clean_paragraph = re.sub(r'[^\w\s]', '', text)
print(clean_paragraph)
```

```python
# Initializing dict() constructor:
counts= dict()
# Storing the cleaned text in a list:
words= clean_paragraph.split()
print(words)
```

### 1st Method: Counting the frequency of words using Dictionary:

```python
start_time2 = datetime.now() # Check the excution time

# for loop starts from here:
for word in words:
  # for every word in words store it in dictionary counts with words as the key and number of occurences as the value:
  counts[word]= counts.get(word,0)+1
print(counts)

# calculating the end time:
end_time2 = datetime.now()
print('Duration: {}'.format(end_time2 - start_time2))
```

```python
# Getting the word with higher number of occurrences in dictionary 'counts':

bigcount= None
bigword= None

for word, count in counts.items():
  if bigcount is None or count> bigcount:
    bigword=word
    bigcount=count
print(bigword, bigcount)
```

### Method 2: Counting the frequency of words using python's Counter():

```python
start_time = datetime.now()

# Passing 'words' as argument for Counter()
count_dict= Counter(words)
print(count_dict)

# Calculate the total execution time for Counter()
end_time = datetime.now()
print('Duration: {}'.format(end_time - start_time))
```

### Plotting histogram to visualize the frequencies:

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(39, 10))
values = count_dict.keys()
counts = count_dict.values()
chart = plt.bar(values, counts)
#chart= sns.barplot(x=values, y= counts)
plt.xticks(rotation=90)  # Rotate x-axis labels by 45 degrees
plt.xlabel('Values')
plt.ylabel('Counts')
plt.title('Histogram')
plt.show()
```

```python
plt.figure(figsize=(39, 10))
sorted_data = sorted(count_dict.items(), key=lambda x: x[1], reverse=True)
values = [item[0] for item in sorted_data]
counts = [item[1] for item in sorted_data]
chart = sns.barplot(x=values, y=counts, palette="rocket")
plt.xticks(rotation=90)
plt.xlabel('Values')
plt.ylabel('Counts')
plt.title('Histogram')
plt.show()
```
